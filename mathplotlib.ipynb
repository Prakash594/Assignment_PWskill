Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
#Scatter Plot: Ideal for visualizing the relationship between 
# two continuous variables by displaying individual data points as markers on a two-dimensional plane.
import numpy as np
from matplotlib import pyplot as plt
np.random.seed(3) 
x = 3 + np.random.normal(0, 2, 50) 
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.

plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
import matplotlib.pyplot as plt
import numpy as np
# For_line_1: 
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1= np.array([0, 100, 200, 300, 400, 500])
# For_line_2: 
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
# For_line_3: 
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
# For_line_4: 
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x1, y1, color='blue')
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2, color='red')
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3, color='black')
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4, color='pink')
plt.title('Line 4')

plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


```python
# A box plot, also known as a box-and-whisker plot, is a graphical representation of 
# the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum. It provides a concise visual summary of the central tendency, dispersion, 
# and skewness of the dataset, along with identifying potential outliers.


box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1,box2]

plt.boxplot(data)
plt.show()
```
