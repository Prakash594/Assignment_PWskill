```python
#activeCount()->
'''
In Python's threading module, the activeCount() method is used to get the number of currently active threads in the program.

The method returns the number of Thread objects that are currently in the "running" state, i.e., 
threads that have started executing but have not yet completed.
This can be useful for monitoring the status of threads and ensuring that 
a program is not creating an excessive number of threads that could lead to performance issues.
'''
#example  how to use activeCount() 
import threading

# create a new thread
t = threading.Thread(target=my_function)
t.start()

# get the number of active threads
num_active_threads = threading.activeCount()

print(f"There are currently {num_active_threads} active threads.")

#currentThread()->
'''
In Python's threading module, the currentThread() method is used to get a reference to the currently executing thread.
The method returns a Thread object that represents the thread from which it was called. 
This can be useful for getting information about the current thread, such as its name, ID, and other attributes.
'''
#example of how to use currentThread() 
import threading

def my_function():
    current_thread = threading.currentThread()
    print(f"Running in thread: {current_thread.getName()}")

# create a new thread
t = threading.Thread(target=my_function)
t.start()

#enumerate()->
'''
In Python, enumerate() is a built-in function that takes an iterable (such as a list or a tuple) as 
its argument and returns an iterator that produces pairs of (index, element) for each element in the iterable.
The first element of each pair is the index of the element in the iterable, starting at 0,
and the second element is the actual value of the element.
'''
#example of how to use enumerate()->
my_list = ["apple", "banana", "cherry"]

for index, value in enumerate(my_list):
    print(f"The element at index {index} is {value}.")
```

Q4. write a python program to create two threads. Thread one must print the list of squares and thread
two must print the list of cubes

```python
import threading

def print_squares():
    for i in range(1, 11):
        print(f"{i} squared is {i*i}")

def print_cubes():
    for i in range(1, 11):
        print(f"{i} cubed is {i*i*i}")

# create two new thread objects and start them
t1 = threading.Thread(target=print_squares)
t2 = threading.Thread(target=print_cubes)
t1.start()
t2.start()

# wait for both threads to finish
t1.join()
t2.join()

print("Main program finished.")
```

Q5. State advantages and disadvantages of multithreading

```python
#advantages
'''
1.Improved performance: Multithreading allows a program to perform multiple tasks concurrently.
This can lead to faster execution times and improved performance, especially on systems with multiple processors or CPU cores.

2.Better resource utilization: Multithreading can help to make better use of system resources, 
such as CPU time, memory, and I/O bandwidth, by allowing multiple threads to share these resources more efficiently.

3.Improved responsiveness: Multithreading can help to improve the responsiveness of a
program or system by allowing user interface and other tasks to be performed in the background while 
the main program continues to run.

4.Simplified program design: Multithreading can simplify the design of complex programs by allowing
different parts of the program to be implemented as separate threads that can communicate and synchronize with each other

'''

#disadvantages

'''
1.Complexity: Multithreading introduces complexity into the code, making it harder to design, implement, and debug.

2.Synchronization: When multiple threads access shared resources, there is a risk of race conditions and deadlocks
that can cause the program to behave unpredictably or crash.

3.Overhead: Creating and managing threads incurs overhead, which can reduce performance and increase resource consumption.

4.Debugging: Finding and fixing bugs in multithreaded code can be more difficult 
than in single-threaded code, especially when bugs occur intermittently.

5.Scalability: Multithreading may not necessarily lead to better performance,
as the overhead of managing threads may outweigh any potential benefits, 
especially on systems with a limited number of processors.

'''
```

Q6. Explain deadlocks and race conditions.

```python
#Deadlocks 
'''
A deadlock occurs when two or more threads are blocked and unable to proceed because they are waiting for
each other to release resources. For example, if Thread 1 holds a lock on Resource A and is waiting for Resource B,
and Thread 2 holds a lock on Resource B and is waiting for Resource A, then both threads will be blocked and unable to proceed, 
causing a deadlock. Deadlocks can be difficult to detect and resolve, and can lead to a program becoming unresponsive or crashing.
'''

#Race
'''
A race condition occurs when the behavior of a program depends on the relative timing of two or 
more threads accessing shared resources, and the outcome of the program is unpredictable. 
For example, if Thread 1 and Thread 2 both access a shared variable X, 
and Thread 1 reads X while Thread 2 is in the process of updating X, 
then the behavior of the program will depend on the relative timing of the two threads, 
and may be unpredictable. Race conditions can lead to bugs that are difficult to detect and reproduce,
and can cause programs to behave in unexpected ways.
'''
```
